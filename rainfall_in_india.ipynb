```python
# import dependencies
import pandas as pd
```

```python
# read in csv for rainfall data
rainfall_df = pd.read_csv("rainfall in india 1901-2015-FILTERED.csv")
rainfall_df
```

```python
# filter data for only 2015
rainfall_2015_df = rainfall_df[rainfall_df['YEAR'] == 2015]
rainfall_2015_df.head()
```

```python
# dropped all but the 'SUBDIVISION' and 'ANNUAL' columns
annual_rainfall_2015_df = rainfall_2015_df[['SUBDIVISION', 'ANNUAL']]
annual_rainfall_2015_df
```

```python
# read cuisine csv
indian_food_df = pd.read_csv("Indian_Food_Cuisine.csv")
indian_food_df['state'] = indian_food_df['state'].str.upper()
indian_food_df.head()
```

```python
# read in state codes
state_codes_df = pd.read_excel("State codes.xlsx")
state_codes_df.head()
```

```python
# merge the indian_food_df and state_codes_df dataframes on state
indian_food_codes_df = indian_food_df.merge(state_codes_df, left_on="state", right_on="STATE", how='inner')
indian_food_codes_df
```

```python
# drop the duplicate column "STATE" after merge
indian_food_codes_df = indian_food_codes_df.drop("STATE", axis = 1)
indian_food_codes_df
```

```python
# rename "TWO-LETTER CODE" column to "State Code"
indian_food_codes_df = indian_food_codes_df.rename(columns={"TWO-LETTER CODE":"State Code"})
indian_food_codes_df
```

```python
# merge annual_rainfall_2015_codes_df and state_codes_df on "SUBDIVISION" and "STATE", respectively
annual_rainfall_2015_codes_df = annual_rainfall_2015_df.merge(state_codes_df, left_on="SUBDIVISION", right_on="STATE", how='inner')
annual_rainfall_2015_codes_df
```

```python
# drop the duplicate state name "SUBDIVISION" column
annual_rainfall_2015_codes_df = annual_rainfall_2015_codes_df.drop("SUBDIVISION", axis = 1)
annual_rainfall_2015_codes_df
```

```python
# indian_food_rainfall_codes_df = indian_food_codes_df and annual_rainfall_2015_codes_df on "State Code" and "TWO-LETTER CODE", respectively
indian_food_rainfall_codes_df = indian_food_codes_df.merge(annual_rainfall_2015_codes_df, left_on="State Code", right_on="TWO-LETTER CODE", how = "left")
indian_food_rainfall_codes_df
```

```python
# drop the "STATE" and "TWO-LETTER CODE" columns
indian_food_rainfall_codes_df = indian_food_rainfall_codes_df.drop(["STATE", "TWO-LETTER CODE"], axis =1)
indian_food_rainfall_codes_df
```

```python
# rename 'ANNUAL' to 'Annual Rainfall (mm)'
indian_food_rainfall_codes_df = indian_food_rainfall_codes_df.rename(columns={'ANNUAL': 'Annual Rainfall (mm)'})
indian_food_rainfall_codes_df
```
